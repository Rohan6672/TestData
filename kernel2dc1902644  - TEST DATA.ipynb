```python
import numpy as np  #imports numpy and names it np
import pandas as pd #imports pandas as pd
import matplotlib.pyplot as plt  #imports matplotlib.pyplot as plt
import seaborn as sns  #imports seaborn as sns 
```

```python
#imports the data set
test_df = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
```

```python
#counts the number of null values in each column 
#uses mathplotlib that makes a frame that is 16X8 inches
#get all the values on the x-axis and counting all the number of nulls on the y-axis
#set the length of each bar to the number of nulls in that column
null_counts = test_df.isnull().sum()
plt.figure(figsize=(16,8))
plt.xticks(np.arange(len(null_counts))+0.5,null_counts.index,rotation='vertical')
plt.ylabel('fraction of rows with missing data')
plt.bar(np.arange(len(null_counts)),null_counts)
```

```python
#counts the number of nulls in each column 
test_df.isnull().sum()
```

```python
# Display the number of non null values in each column
# Also tells us the data type in each column
test_df.info()
```

```python
#fills up all the nulls with the mean value for column with integer data types
#fills up all the nulls with the mode value for column with non integer data types

test_df['LotFrontage']=test_df['LotFrontage'].fillna(test_df['LotFrontage'].mean()) 
test_df['MSZoning']=test_df['MSZoning'].fillna(test_df['MSZoning'].mode()[0])
test_df['SaleType']=test_df['SaleType'].fillna(test_df['SaleType'].mode()[0])
```

```python
#returns the number of zeros to check to see if line 9 worked
test_df.isnull().sum()
```

```python
#gives us the size of the data frame by telling the number of rows,columns
test_df.shape
```

```python
#drops the column Alley since it has a lot of null values
#checking if it dropped 
test_df.drop(['Alley'],axis=1,inplace=True)
test_df.head()
```

```python
#checking if the number of colums decreased by one 
test_df.shape
```

```python
#replaces the null values of BsmtCond with the mode of the column
#replaces the null values of BsmtQual with the mode of the column
#replaces the null values of FireplaceQu with the mode of the column
#replaces the null values of GarageType with the mode of the column
#replaces the null values of GarageFinish with the mode of the column
#replaces the null values of GarageQual with the mode of the column
#replaces the null values of GarageCond with the mode of the column
#replaces the null values of BsmtExposure with the mode of the column
#replaces the null values of BsmtFinType1 with the mode of the column
#replaces the null values of BsmtCond with the mode of the column
#replaces the null values of BsmtCond with the mode of the column
test_df['BsmtCond']=test_df['BsmtCond'].fillna(test_df['BsmtCond'].mode()[0])
test_df['BsmtQual']=test_df['BsmtQual'].fillna(test_df['BsmtQual'].mode()[0])
test_df['FireplaceQu']=test_df['FireplaceQu'].fillna(test_df['FireplaceQu'].mode()[0])
test_df['GarageType']=test_df['GarageType'].fillna(test_df['GarageType'].mode()[0])
test_df['GarageFinish']=test_df['GarageFinish'].fillna(test_df['GarageFinish'].mode()[0])
test_df['GarageQual']=test_df['GarageQual'].fillna(test_df['GarageQual'].mode()[0])
test_df['GarageCond']=test_df['GarageCond'].fillna(test_df['GarageCond'].mode()[0])
test_df['BsmtExposure']=test_df['BsmtExposure'].fillna(test_df['BsmtExposure'].mode()[0])      
test_df['BsmtFinType1']=test_df['BsmtFinType1'].fillna(test_df['BsmtFinType1'].mode()[0])  
test_df['BsmtFinType2']=test_df['BsmtFinType2'].fillna(test_df['BsmtFinType2'].mode()[0])
test_df['FireplaceQu']=test_df['FireplaceQu'].fillna(test_df['FireplaceQu'].mode()[0])
```

```python
#gives the info of the data 
test_df.info()
```

```python
# Fill missing values of GarageYrBuilt with corresponding values from YearBuilt
test_df['GarageYrBlt'] = test_df['GarageYrBlt'].fillna(test_df['YearBuilt'])
```

```python
# axis=1 when you want to refer to a column
# drop column 'Id'
test_df.drop(['PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
test_df.drop(['Id'],axis=1,inplace=True)
```
